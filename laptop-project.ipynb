# STORAGE PROCESSING :

```python
df["Storage"].unique()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(data=df,x="Storage")
plt.show()
```

```python
df["Storage"].value_counts()
```

```python
df.dropna(inplace=True)
```

```python
def mb_convert(s):
    print(s)
    num=''
    if 'GB' in s:
        for i in s:
            if i.isdigit():
                num+=i
        return int(num)*1024
    elif 'TB' in s:
        for i in s:
            if i.isdigit():
                num+=i
        return int(num)*(1024**2)
```

```python
df['Storage']=df['Storage'].apply(mb_convert)
# df['Storage']=df['Storage'].astype(int)
# # df['Storage']
```
